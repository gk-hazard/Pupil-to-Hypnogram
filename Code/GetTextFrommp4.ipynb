## acquire timestamp

reference: https://scrapbox.io/nwtgck/Python%E3%81%A7%E3%83%95%E3%83%AC%E3%83%BC%E3%83%A0%E3%81%94%E3%81%A8%E3%81%AB%E3%82%BF%E3%82%A4%E3%83%A0%E3%82%B9%E3%82%BF%E3%83%B3%E3%83%97%E3%82%92%E5%8F%96%E5%BE%97%E3%81%99%E3%82%8B%E6%96%B9%E6%B3%95

```python
import cv2

#Change the path below
video_path = "../Data/Inputs/SNT267_0806.mp4"
```

```python
def get_timestamps(video_path):
  cap = cv2.VideoCapture(video_path)
  timestamps = [cap.get(cv2.CAP_PROP_POS_MSEC)]
  while cap.isOpened():
    frame_exists, curr_frame = cap.read()
    if not frame_exists:
      break
    timestamp = cap.get(cv2.CAP_PROP_POS_MSEC)
    timestamps.append(timestamp)
  cap.release()
  return timestamps
```

```python
timestamp = get_timestamps(video_path)
```
